```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
html = requests.get('https://gist.github.com/paulmillr/2657075').text
soup = BeautifulSoup(html)
```

```python
soup.select('td')[632].text
```

```python
dict = {}
for i in range (0,1000,4):
    liste = []
    liste.append(soup.select('td')[i].text)
    liste.append(soup.select('td')[i].find('a').attrs['href'])
    dict[i] = liste
```

```python
df = pd.DataFrame(list(dict.items()))
df.columns = ['Numéro','n_p']
df = df.drop(columns="Numéro")
```

```python
df
```

```python
df['nom'] = df.n_p.str[0]
```

```python
df['nom'] = df['nom'].str.split(pat=" ", n=1, expand=True)
```

```python
df
```

```python
import statistics as stat

def calc_mean(username):
    url = f'https://github.com/{username}'
    print(url)
    html1 = requests.get(url).text
    soup1 = BeautifulSoup(html1)
    st = soup1.select('a[href*="stargazers"]')
    if(len(st) > 0):
        stars = [float(star.get_text().strip()[:-1])*1000 if 'k' in star.get_text().strip() else float(star.get_text().strip()) for star in st]
        moyenne = stat.mean(stars)
        print(moyenne)
        return moyenne
    else:
        return 0
```

```python
calc_mean(df['nom'][1])
```

```python
j = 0
df['mean'] = "decl"
for i in df['nom']:
    df['mean'][j] = calc_mean(i)
    j = j + 1
```
